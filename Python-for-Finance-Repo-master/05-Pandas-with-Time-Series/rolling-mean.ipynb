```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('time_data/walmart_stock.csv', parse_dates=True, index_col='Date')
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
df.rolling(7).mean().head(14)
```

```python
df['Open'].plot()
df.rolling(window=90).mean()['Close'].plot(figsize=(16,6))
```

```python
df['Close 30 day MA'] = df['Close'].rolling(window=30).mean()
df[['Close 30 day MA', 'Close']].plot(figsize=(16,6))
```

```python
df['Close'].expanding().mean().plot(figsize=(16,6))
```

```python
# Close 20 Moving Average
df['Close: 20 Day Mean'] = df['Close'].rolling(window=20).mean()

# Upper = 20 MA + 2 Std(20 Days)
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(window=20).std())

# Lower = 20 MA - 2 Std(20 Days)
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(window=20).std())
# Close

df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].plot(figsize=(16,6))
```
