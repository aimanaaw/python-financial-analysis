```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional Method
plt.plot(x,y)
#plt.show()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object oriented method
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1,0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

# for ax in axes:
#     ax.plot(x,y)
    
axes[0].plot(x,y)
axes[1].plot(y,x)
# axes.plot(x,y)

plt.tight_layout()
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2))

# ax = fig.add_axes([0,0,1,1])
# ax.plot(x,y)
axes[0].plot(x,y)
axes[0].set_xlabel('x')
axes[1].plot(y,x)
axes[1].set_ylabel('y')

plt.tight_layout()
```

```python
fig.savefig('my_picture.pdf')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.legend(loc=1)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,2,2])

axes.plot(x,y,color='red', lw=0.5, linestyle='-.',marker='+', markersize=10)
```
